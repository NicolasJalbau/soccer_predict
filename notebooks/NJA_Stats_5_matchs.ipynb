```python
import pandas as pd
import numpy as np
```

```python
stat = pd.read_csv('../raw_data/teamstats.csv')
```

```python
import datetime
stat = stat.sort_values(by='date')
teamID = 164
match_date = datetime.datetime(2015,1,9)
n_match = 5
```

```python
team_stat = stat[stat['teamID'] == teamID]
```

```python
team_stat['date'] = pd.to_datetime(team_stat['date'])
team_stat[team_stat['date'] < match_date]
```

```python
stat_5_match = team_stat[team_stat['date'] < match_date].tail(n_match).sum()
```

```python
stat_5_match.drop(['gameID','season'],inplace=True)
```
