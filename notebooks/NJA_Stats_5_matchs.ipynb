```python
def get_stats_previous_matchs(team_id,match_date,n_match=5):
    #root_dir = os.path.dirname(os.path.dirname(__file__))
    csv_path = os.path.join('..', 'raw_data','teamstats.csv')
    stat = pd.read_csv(csv_path).copy()
    stat['date'] = pd.to_datetime(stat['date'])
    stat = stat.sort_values(by='date')
    team_stat = stat[stat['teamID'] == teamID]
    stat_5_match = team_stat[team_stat['date'] < match_date].tail(n_match).sum()
    stat_5_match.drop(['gameID','season'],inplace=True)
    return pd.DataFrame(stat_5_match.values.reshape(1,-1),columns=stat_5_match.index)
```

```python
stat_marseille = get_stats_previous_matchs(164,datetime.datetime(2014,8,9))
```
